```
from pyspark.sql import SparkSession
spark= SparkSession.builder.master("local[*]").appName('Customers').getOrCreate()
from pyspark import SparkConf, SparkContext
import collections
conf= SparkConf().setMaster("local").setAppName("word couting")
sc = SparkContext.getOrCreate(conf=conf)
```

```
from pyspark.ml.regression import LinearRegression
from pyspark import SparkContext
from pyspark.sql import functions as f
```

```
content =sc.textFile("text.TXT")
```

```
counts = content.flatMap(lambda line: line.split(" ")) \
             .map(lambda word: (word, 1)) \
             .reduceByKey(lambda a, b: a + b)
```

```
counts.collect()    # -------> prints all lines
```

```
columns=["word","count"]
```

```
df = spark.createDataFrame(data = counts, schema = columns)
df.printSchema()
df.show(truncate=False)
```

```
df.sort("count").show(truncate=False)
```

```
df.orderBy(f.col("count")).show(truncate=False)
```

```
df.orderBy(f.col("count").desc()).show(truncate=False)
```
